```python
"""make a recipe as a list column"""

def recipe_splitter(x):
    return x.split(";")
    
df['recipe_list'] = df['recipe'].apply(recipe_splitter)
df.head()
```

```python
def one_hot_encode_list_column(df, column_name):
    """
    One-hot encodes a column containing lists in a DataFrame.

    Args:
        df: The DataFrame containing the column to encode.
        column_name: The name of the column containing lists.

    Returns:
        The DataFrame with the original column and the one-hot encoded columns.
    """
    mlb = MultiLabelBinarizer()
    one_hot_encoded = mlb.fit_transform(df[column_name])
    one_hot_df = pd.DataFrame(one_hot_encoded, columns=mlb.classes_)
    return pd.concat([df, one_hot_df], axis=1)


df = one_hot_encode_list_column(df, 'recipe_list')
df.head()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu, ks_2samp

# Define your recipe groups
recipe_groups = ['Best Ranked', 'Worst Ranked', 'Overlap', 'Other']

# Prepare a DataFrame to hold p-values for both tests
p_values_mw = pd.DataFrame(index=recipe_groups, columns=recipe_groups)
p_values_ks = pd.DataFrame(index=recipe_groups, columns=recipe_groups)

# Iterate over all pairs of recipe groups to calculate p-values
for i, group1 in enumerate(recipe_groups):
    for j, group2 in enumerate(recipe_groups):
        if i >= j:
            continue
        data1 = df[df['recipe_group'] == group1]['recipe_diff']
        data2 = df[df['recipe_group'] == group2]['recipe_diff']
        
        # Mann-Whitney U test
        mw_stat, mw_p = mannwhitneyu(data1, data2, alternative='two-sided')
        p_values_mw.loc[group1, group2] = mw_p
        p_values_mw.loc[group2, group1] = mw_p
        
        # Kolmogorov-Smirnov test
        ks_stat, ks_p = ks_2samp(data1, data2)
        p_values_ks.loc[group1, group2] = ks_p
        p_values_ks.loc[group2, group1] = ks_p

# Convert DataFrame values to float for plotting
p_values_mw = p_values_mw.astype(float)
p_values_ks = p_values_ks.astype(float)

# Plot the heatmap for Mann-Whitney U test p-values
plt.figure(figsize=(10, 8))
sns.heatmap(p_values_mw, annot=True, cmap='viridis', cbar_kws={'label': 'p-value'})
plt.title('Mann-Whitney U Test p-values')
plt.xlabel('Recipe Group')
plt.ylabel('Recipe Group')
plt.show()

# Plot the heatmap for Kolmogorov-Smirnov test p-values
plt.figure(figsize=(10, 8))
sns.heatmap(p_values_ks, annot=True, cmap='viridis', cbar_kws={'label': 'p-value'})
plt.title('Kolmogorov-Smirnov Test p-values')
plt.xlabel('Recipe Group')
plt.ylabel('Recipe Group')
plt.show()
```
